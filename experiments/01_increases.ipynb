```python
from pathlib import Path
import pandas as pd
```

```python
states = ["Alabama","Alaska","Arizona","Arkansas","California","Colorado",
  "Connecticut","Delaware","Florida","Georgia","Hawaii","Idaho","Illinois",
  "Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland",
  "Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana",
  "Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York",
  "North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania",
  "Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah",
  "Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"]
```

```python
#hacky shortcut to get all the paths to data folders
path_new = Path('../2020_05_04')
path_old = Path('../2020_04_28.02')
```

```python
df_new = pd.read_csv(path_new/"hospitalization_all_locs.csv")
df_old = pd.read_csv(path_old/"hospitalization_all_locs.csv")
```

```python
df_old
```

```python
df_old.columns
```

```python
df_old_preds = df_old[(df_old["date"] == "2020-08-04")&(df_old["location_name"].isin(states))][['location_name', 'totdea_mean']]
```

```python
df_new_preds = df_new[(df_new["date"] == "2020-08-04")&(df_new["location_name"].isin(states))][['location_name', 'totdea_mean']]
```

```python
df_old_preds.head(10)
```

```python
df_both = pd.merge(df_old_preds, df_new_preds, on='location_name')
```

```python
df_both = df_both.rename(columns={"totdea_mean_x":"total_deaths_old", "totdea_mean_y":"total_deaths_new"})
```

```python
df_both.head(5)
```

```python
df_both["increase"] = df_both['total_deaths_new'] - df_both['total_deaths_old']
```

```python
df_both["increase_pct"] = 100 * (df_both['total_deaths_new'] - df_both['total_deaths_old'])/df_both['total_deaths_old']
```
